```python
import numpy as np
directions=np.array([[1,0],[0,1],[-1,0],[0,-1]])
```

```python
area=[]
linecount = 0
with open('input_16.txt') as f:
    for line in f:
        area.append(list(line.strip()))
maparr=np.array(area)
dimx=maparr.shape[0]
dimy=maparr.shape[1]
print(dimx,dimy,maparr)
```

```python
#Find Start
[startx,starty]=np.where(maparr=="S")
start=np.array([startx[0],starty[0]])
print(start)
startdir=1
#Find target
[targetx,targety]=np.where(maparr=="E")
target=np.array([targetx[0],targety[0]])
print(target)
```

```python
visited=np.full((dimx,dimy,4),-1)
locs=np.array([start])
oris=np.array([startdir])
costs=np.array([0])
visited[start[0],start[1],startdir]=0
while len(locs)>0:
    #print('New round')
    newlocs=[]
    newdirs=[]
    newcosts=[]
    for loc,ori,cost in zip(locs,oris,costs):
        #print('Location', loc,'Ori',ori,'Cost',cost)
        for dir in range(ori-1,ori+2):
            dir %=4
            #print ('direction', dir)
            if maparr[loc[0]+directions[dir][0],loc[1]+directions[dir][1]]!='#':
                prevcost=visited[loc[0]+directions[dir][0],loc[1]+directions[dir][1],dir]
                newcost=cost+ (1 if (dir==ori) else 1001)
                #print('Try',loc[0]+directions[dir][0],loc[1]+directions[dir][1],prevcost,newcost)
                if prevcost==-1 or prevcost>newcost:
                    #print('Success')
                    newlocs.append(loc+directions[dir])
                    newdirs.append(dir)
                    newcosts.append(newcost)
                    visited[loc[0]+directions[dir][0],loc[1]+directions[dir][1],dir]=newcost
        #locs=np.array(np.unique(climb,axis=0))
        locs=np.array(newlocs)
        oris=np.array(newdirs)
        costs=np.array(newcosts)
print('Final',visited[target[0],target[1],:],np.sort(visited[target[0],target[1],:]))
```

```python
min(visited[target[0],target[1],visited[target[0],target[1],:]>-1])
bestdirs=np.where(visited[target[0],target[1],:]==min(visited[target[0],target[1],visited[target[0],target[1],:]>-1]))
bestdirs=bestdirs[0]
bestcost=min(visited[target[0],target[1],visited[target[0],target[1],:]>-1])
```

```python
seats=np.zeros_like(maparr,dtype=int)
locs=np.array([target])
#oris=np.array(bestdirs)
curcosts=np.array([visited[target[0],target[1],bestdirs[0]]])
print(curcosts)
seats[target[0],target[1]]=1
while(len(locs)>0):
    newlocs=[]
    newcosts=[]
    for loc,curcost in zip(locs,curcosts):
        #print('Location', loc, 'Orientation',ori,'Curcost',curcost)
        for dir in range(4):
            print ('direction', dir)
            if maparr[loc[0]+directions[dir][0],loc[1]+directions[dir][1]]!='#':
                prevcosts=visited[loc[0]+directions[dir][0],loc[1]+directions[dir][1],:]
                #print('Try',loc[0]+directions[dir][0],loc[1]+directions[dir][1],prevcosts,curcost)
                for prevcost in prevcosts:
                    if prevcost!=-1 and (curcost-prevcost==1001 or curcost-prevcost==1):
                        #print('Success', prevcost)
                        newlocs.append(loc+directions[dir])
                        newcosts.append(prevcost)
                        seats[loc[0]+directions[dir][0],loc[1]+directions[dir][1]]=1
        #locs=np.array(np.unique(climb,axis=0))
        locs=np.array(newlocs)
        curcosts=np.array(newcosts)
print('Final',seats)
```
