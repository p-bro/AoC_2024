```python
import numpy as np
import pandas as pd
import re
```

```python
sum=0
with open('input_03.txt') as f:
    for line in f:
        muls=re.findall("mul\([0-9]+,[0-9]+\)",line)
        for mul in muls:
            fact=(list(map(int,re.split(",",re.findall("[0-9,]+",mul)[0]))))
            if(fact[0]<1000 and fact[1]<1000):
                sum += fact[0]*fact[1]
print(sum)
```
