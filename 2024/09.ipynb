```python
linecount = 0
with open('input09b.txt') as f:
    for line in f:
        code=list(line.strip())
print(code)
```

```python
fileid=0
file=1
blocks=[]
chksum=0
for instr in code:
    if file:
        for i in range(int(instr)):
            blocks.append(fileid)
        fileid +=1
        file=0
    else: #space
        for i in range(int(instr)):
            blocks.append(-1)
        file=1
#print(blocks)
print(sum(blocks))
print(len(blocks))
lastblock=len(blocks)-1
for [blockct,block] in enumerate(blocks):
    if block>-1:
        chksum += block*blockct
        #print (block*blockct)
        continue
    while (blocks[lastblock]==-1):
        lastblock -=1
    if lastblock<=blockct:
        break
    #empty cell, pull last one
    blocks[blockct]=blocks[lastblock]
    blocks[lastblock]=-1
    chksum+=blocks[blockct]*blockct
    #print(blocks[blockct]*blockct)
#print(blocks)
print(chksum)
chksum=0
for [blockct,block] in enumerate(blocks):
    if block>-1:
        chksum += block*blockct

print(chksum)
print(sum(blocks))
```
