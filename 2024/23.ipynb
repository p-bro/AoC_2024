```python
import numpy as np
import networkx as nx
import re
```

```python
connections=[]
with open('input_23.txt') as f:
    for line in f:
        m=re.match('([a-z]+)-([a-z]+)',line)
        connections.append((m.group(1),m.group(2)))
print(connections)
```

```python
G=nx.Graph()
G.add_edges_from(connections)
nodes=list(G.nodes)
bossnodes=[node for node in nodes if node[0]=='t']
print(nodes,bossnodes)
```

```python
all_cliques= nx.enumerate_all_cliques(G)
triad_cliques=[x for x in all_cliques if len(x)==3 ]
```

```python
boss_triads=[triad for triad in triad_cliques if triad[0] in bossnodes or triad[1] in bossnodes or triad[2] in bossnodes]
```

```python
print(len(boss_triads),boss_triads)
```

```python
cliques=nx.find_cliques(G)
for clique in nx.find_cliques(G):
    print(clique)
#cliques = [item for item in res]
cliques = sorted(cliques, key=lambda item: -len(item))
print (cliques[0])
sorted(cliques[0])
for node in sorted(cliques[0]):
    print(node,end=',')
```
