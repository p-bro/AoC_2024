```python
def numkey(cur,dest):
    if cur==dest:
        return ''
    seq=cur+dest
    match seq:
        case '47'|'58'|'69'|'14'|'25'|'36'|'02'|'A3':
            return '^'
        case '74'|'85'|'96'|'41'|'52'|'63'|'20'|'3A':
            return 'v'
        case '17'|'28'|'39'|'05'|'A6':
            return '^^'
        case '71'|'82'|'93'|'50'|'6A':
            return 'vv'
        case '08'|'A9':
            return '^^^'
        case '80'|'9A':
            return 'vvv'
        case '78'|'89'|'45'|'56'|'12'|'23'|'0A':
            return '>'
        case '87'|'98'|'54'|'65'|'21'|'32'|'A0':
            return '<'
        case '79'|'46'|'13':
            return '>>'
        case '97'|'64'|'31':
            return '<<'
        case '48'|'59'|'15'|'26'|'03':
            return '>^'
        case '84'|'95'|'51'|'62'|'30':
            return '<v'
        case '57'|'68'|'24'|'35'|'A2':
            return '<^'
        case '01':
            return '^<'
        case '75'|'86'|'42'|'53'|'2A':
            return 'v>'
        case '10':
            return '>v'
        case '18'|'29'|'06':
            return '>^^'
        case '81'|'92'|'60':
            return '<vv'
        case '27'|'38'|'A5':
            return '<^^'
        case '04':
            return '^^<'
        case '72'|'83'|'5A':
            return 'vv>'
        case '40':    
            return '>vv'
        case '49'|'16':
            return '>>^'
        case '94'|'61':
            return '<<v'
        case '67'|'34':
            return '<<^'           
        case 'A1':
            return '^<<'
        case '76'|'43':
            return 'v>>'            
        case '1A':
            return '>>v'
        case '19':
            return '>>^^'
        case '91':
            return '<<vv'
        case '37' :
            return '<<^^'           
        case 'A4':
            return '^^<<'
        case '73':
            return 'vv>>'
        case '4A':
            return '>>vv'
        case 'A8':
            return '<^^^'
        case '07':
            return '^^^<'
        case '8A':
            return 'vvv>'
        case '70':
            return '>vvv'
        case '09':
            return '>^^^'
        case '90':
            return 'vvv>'
        case 'A7':
            return '^^^<<'
        case '7A':
            return '>>vvv'
        case _:
            raise ValueError('Codes undefined',seq)

def dirkey (cur,dest):
    if cur==dest:
        return ''
    seq=cur+dest        
    match seq:
        case '^A'|'<v'|'v>':
            return '>'
        case 'A^'|'v<'|'>v':
            return '<'
        case '<>':
            return '>>'
        case '><':
            return '<<'
        case 'v^'|'>A':
            return '^'
        case '^v'|'A>':
            return 'v'
        case '<^':
            return '>^'
        case 'vA':
            return '^>'
        case '^<':
            return 'v<'
        case 'Av':
            return '<v'
        case '>^':
            return '<^'        
        case '^>':
            return 'v>'
        case '<A':       
            return '>>^'
        case 'A<':       
            return 'v<<'
        case _:
            raise ValueError('Codes undefined:',seq)
            
```

```python
codes=[]
linecount = 0
with open('input_21.txt') as f:
    for line in f:
        codes.append(line.strip())
print(codes)
```

```python
nummap={}
for i in '0123456789A':
    for j in '0123456789A':
        print (i,j,numkey(i,j))
        if len(numkey(i,j))==1:
            nummap[i+numkey(i,j)]=j
dirmap={}
for i in '<>^vA':
    for j in '<>^vA':
        if len(dirkey(i,j))==1:
            dirmap[i+dirkey(i,j)]=j
        print (i,j,dirkey(i,j))
print(nummap)
print(dirmap)
```
