```python
import numpy as np
import pandas as pd
```

```python
safe=0
with open('input_02.txt') as f:
    for line in f:
        condition=np.array(list(map(int,line.strip().split())))
        print(condition)
        status=1
        if (condition.size>1):
            if condition[1]>condition[0]:
                incr=1
            elif condition[1]<condition[0]:
                incr=-1
            else:
                print(condition," is unsafe")
                incr=0
                status=0
            #    continue
            for i in range(1,np.size(condition)):
                if incr*(condition[i]-condition[i-1])<=0 or incr*(condition[i]-condition[i-1])>3:
                    print(condition,"is unsafe")
                    status=0
                    break
            if status==1:
                print(condition," is safe")
                safe += 1
            else: #Unsafe
                for skip in range(np.size(condition)):
                    status=1
                    short=np.delete(condition,skip)
                    if (short.size>1):
                        if short[1]>short[0]:
                            incr=1
                        elif short[1]<short[0]:
                            incr=-1
                        else:
                            print(short," is unsafe")
                            incr=0
                            status=0
                            #continue
                        for i in range(1,np.size(short)):
                            if incr*(short[i]-short[i-1])<=0 or incr*(short[i]-short[i-1])>3:
                                print(short,"is unsafe at ",i)
                                status=0
                                break
                        if status==1:
                            print(short," is safe")
                            safe +=1
                            break

        else:
            raise(ValueError)
print(safe)
```
