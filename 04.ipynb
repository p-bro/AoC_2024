```python
import numpy as np
```

```python
xmas=[]
linecount = 0
with open('input_04.txt') as f:
    for line in f:
        xmas.append(list(line.strip()))
xmasarr=np.array(xmas)
dimx=xmasarr.shape[0]
dimy=xmasarr.shape[1]
print(dimx,dimy)
```

```python
counter=0
for i in range(dimx):
    for j in range(dimy):
        if xmasarr[i,j]=='X':
            if i+3<dimx:
                # Check west
                if xmasarr[i+1,j]=='M' and xmasarr[i+2,j]=='A' and xmasarr[i+3,j]=='S':
                    counter+=1
                #Check southwest
                if j+3<dimy:
                    if xmasarr[i+1,j+1]=='M' and xmasarr[i+2,j+2]=='A' and xmasarr[i+3,j+3]=='S':
                        counter+=1
                #Check northwest
                if j-3>=0:
                    if xmasarr[i+1,j-1]=='M' and xmasarr[i+2,j-2]=='A' and xmasarr[i+3,j-3]=='S':
                        counter+=1
            if i-3>=0:
                # Check east
                if xmasarr[i-1,j]=='M' and xmasarr[i-2,j]=='A' and xmasarr[i-3,j]=='S':
                    counter+=1
                #Check southeast
                if j+3<dimy:
                    if xmasarr[i-1,j+1]=='M' and xmasarr[i-2,j+2]=='A' and xmasarr[i-3,j+3]=='S':
                        counter+=1
                #Check northeast
                if j-3>=0:
                    if xmasarr[i-1,j-1]=='M' and xmasarr[i-2,j-2]=='A' and xmasarr[i-3,j-3]=='S':
                        counter+=1
            if j+3<dimy:
                #check south
                if xmasarr[i,j+1]=='M' and xmasarr[i,j+2]=='A' and xmasarr[i,j+3]=='S':
                    counter+=1
            if j-3>=0:
                #check north
                if xmasarr[i,j-1]=='M' and xmasarr[i,j-2]=='A' and xmasarr[i,j-3]=='S':
                    counter+=1
print(counter)
```
