```python
import numpy as np
import re
```

```python
keys=[]
locks=[]
lc=0
with open('input_25.txt') as f:
    for line in f:
        if len(line)<2: #blank lines separate rules from updates
            lc=0
            continue
        if lc==0:
            if line[0]=="#":
                islock=True
                lock=[list(line.strip())]
            elif line[0]=='.':
                islock=False
                key=[list(line.strip())]
            lc=1
            continue
        lc+=1
        if islock:
            lock.append(list(line.strip()))
        else:
            key.append(list(line.strip()))
        if lc==7:
            if islock:
                lock=np.array(lock)
                code=np.zeros(5,dtype=int)
                for pin in range(5):
                    code[pin]=np.count_nonzero(lock[:,pin]=='#')-1
                locks.append(code)
                print(lock,lock.shape,code)
            else:
                key=np.array(key)
                code=np.zeros(5,dtype=int)
                for pin in range(5):
                    code[pin]=np.count_nonzero(key[:,pin]=='#')-1
                keys.append(code)
                print(key,key.shape,code)
```

```python
count=0
fit=np.full(5,5,dtype=int)
for key in keys:
    for lock in locks:
        if all(key+lock<=fit):
            print('Match:',key,lock)
            count+=1
print(count)
```
