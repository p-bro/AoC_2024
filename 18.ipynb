```python
import numpy as np
import re
directions=np.array([[1,0],[0,1],[-1,0],[0,-1]])
```

```python
dimx=71
dimy=71
numbytes=1024
bytelist=[]
linecount = 0
with open('input_18.txt') as f:
    for line in f:
        m=re.match('([0-9]+),([0-9]+)',line)
        bytelist.append([int(m.group(1)),int(m.group(2))])
bytelist=np.array(bytelist)
maparr=np.full((dimx,dimy),'.',dtype=str)
print(maparr)
for count,block in enumerate(bytelist):
    print(count,block)
    maparr[block[0],block[1]]='#'
    if count==numbytes-1:
        break
maparr[0,0]='S'
maparr[-1,-1]='E'
print(maparr.transpose(),dimx,dimy)
```

```python
#Find Start
[startx,starty]=np.where(maparr=="S")
start=np.array([startx[0],starty[0]])
print(start)
startdir=1
#Find target
[targetx,targety]=np.where(maparr=="E")
target=np.array([targetx[0],targety[0]])
print(target)
```

```python
visited=np.full((dimx,dimy),-1)
locs=np.array([start])
visited[start[0],start[1]]=0
while len(locs)>0:
    #print('New round')
    newlocs=[]
    #newdirs=[]
    #newcosts=[]
    for loc in locs:
        cost=visited[loc[0],loc[1]]
        #print('Location', loc,'Cost',cost)
        for dir in directions:
            #print ('direction', dir)
            if loc[0]+dir[0]>=0 and loc[1]+dir[1]>=0 and loc[0]+dir[0]<dimx and loc[1]+dir[1]<dimy:
                if maparr[loc[0]+dir[0],loc[1]+dir[1]]!='#' and visited[loc[0]+dir[0],loc[1]+dir[1]]==-1:
                    #print('Success')
                    newlocs.append(loc+dir)
                    visited[loc[0]+dir[0],loc[1]+dir[1]]=cost+1
        locs=np.array(np.unique(newlocs,axis=0))
        #locs=np.array(newlocs)
print(visited.transpose())
print('Final',visited[target[0],target[1]])

```
