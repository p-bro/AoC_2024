```python
import numpy as np
directions=np.array([[1,0],[0,1],[-1,0],[0,-1]])
dirs='<>^v'
dirnums='3120'
dirtrans=str.maketrans(dirs,dirnums)
```

```python
area=[]
oplist=[]
linecount = 0
with open('input_15.txt') as f:
    for line in f:
        if len(line)<2:
            break
        area.append(list(line.strip().replace('.','..').replace('@','@.').replace('#','##').replace('O','[]')))
    for line in f:
        oplist.append(list(line.strip()))
maparr=np.array(area)
oparr=np.array(oplist).flatten()
ops=np.char.translate(oparr,dirtrans).astype(int)
dimx=maparr.shape[0]
dimy=maparr.shape[1]
robotx,roboty=np.where(maparr=='@')
robot=np.array([robotx[0],roboty[0]])
print(dimx,dimy,maparr,oparr,ops,robot)
#maparr[robot[0],robot[1]]='.'
for i in range(dimx):
    print(''.join(maparr[i,:]))
```

```python
for op in ops:
    boxcount=0
    pushpos=[]
    allpushpos=[]
    #print ('Try ',robot, op, directions[op], maparr[robot[0],robot[1]])
    pushpos.append([robot[0],robot[1]])
    allpushpos.append([robot[0],robot[1]])
    pushdepth=1
    pushing=1
    wall=0
    while len(pushpos)>0:
        newpushpos=[]
        for push in pushpos:
            #print('Checking at',push)
            if maparr[push[0]+directions[op][0],push[1]+directions[op][1]]=='#': # Wall
                #print('Wall at',push[0]+directions[op][0],push[1]+directions[op][1])
                wall=1
                break #Nothing happens
            if op==3: #left
                if maparr[push[0]+directions[op][0],push[1]+directions[op][1]]==']':
                    #print('Push box left',push[0]+directions[op][0],push[1]+directions[op][1])
                    allpushpos.append([push[0]+directions[op][0],push[1]+directions[op][1]])
                    #allpushpos.append[push[0]+2*directions[op][0],push[1]+2*directions[op][1]]
                    newpushpos.append([push[0]+2*directions[op][0],push[1]+2*directions[op][1]])
            if op==1: #right
                if maparr[push[0]+directions[op][0],push[1]+directions[op][1]]=='[':
                    #print('Push box right',push[0]+directions[op][0],push[1]+directions[op][1])
                    allpushpos.append([push[0]+directions[op][0],push[1]+directions[op][1]])
                    #allpushpos.append[push[0]+2*directions[op][0],push[1]+2*directions[op][1]]
                    newpushpos.append([push[0]+2*directions[op][0],push[1]+2*directions[op][1]])
            if op==2 or op==0:
                #print('Push box up down', op, directions[op])
                if maparr[push[0]+directions[op][0],push[1]+directions[op][1]]==']':
                    newpushpos.append([push[0]+directions[op][0],push[1]+directions[op][1]])
                    newpushpos.append([push[0]+directions[op][0],push[1]-1+directions[op][1]])
                if maparr[push[0]+directions[op][0],push[1]+directions[op][1]]=='[':
                    newpushpos.append([push[0]+directions[op][0],push[1]+directions[op][1]])
                    newpushpos.append([push[0]+directions[op][0],push[1]+1+directions[op][1]])
        if wall==1:
            break
        for newpos in np.unique(newpushpos,axis=0):
            allpushpos.append(newpos)
        pushpos=np.unique(newpushpos,axis=0)
    if not wall:
        for pos in reversed(allpushpos):
            temp=maparr[pos[0]+directions[op][0],pos[1]+directions[op][1]]
            maparr[pos[0]+directions[op][0],pos[1]+directions[op][1]]=maparr[pos[0],pos[1]]
            maparr[pos[0],pos[1]]=temp
        #maparr[robot[0],robot[1]]='.'
        robot=robot+directions[op]
#    for i in range(dimx):
#        print(''.join(maparr[i,:]))
```
