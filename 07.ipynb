```python
# Define addition operation function
def add(a, b):
    return a+b

# Define subtraction operation function
def sub(a, b):
    return a-b

# Define multiplication operation function
def mul(a, b):
    return a*b

# Define division operation function
def div(a, b):
    return a/b
#operators=(add,sub,mul,div)
def con(a,b):
    return a*10**len(str(b))+b
operators=(add,mul,con)
```

```python
sum=0
numoperators=3
with open('input_07.txt') as f:
    for line in f:
        equation=line.strip().split()
#        print (equation)
        result=int(equation[0][:-1])
#        print(result)
        operands=np.array(equation[1:],dtype=int)
#        print(operands)
        numops=operands.size-1
        #create base 4 lists of right length
        for i in range(numoperators**numops):
            num=i
            oparray=[]
            for j in range(numops):
                if i==0:
                    oparray.append(0)
                    continue
                oparray.append(int(num%numoperators))
                num = num // numoperators
            #print(oparray)
            testres=operands[0]
            #print(testres)
            for opnum,op in enumerate(oparray):
            #    print(opnum,op)
                testres=operators[op](testres,operands[opnum+1])
            #    print (testres)
            if testres==result:
                print('match', line.strip(),oparray)
                sum += result
                break
        #for operand in range(1,operands):
        #    for operator in operators:
        #        testres[operand]=operator(testres[operand-1],operands[operand])
print(sum)
```
