```python
import numpy as np
directions=np.array([[1,0],[0,1],[-1,0],[0,-1]])
```

```python
area=[]
linecount = 0
with open('input_20.txt') as f:
    for line in f:
        area.append(list(line.strip()))
maparr=np.array(area)
dimx=maparr.shape[0]
dimy=maparr.shape[1]
print(dimx,dimy,maparr)
maparrorig=maparr.copy()
#Find Start
[startx,starty]=np.where(maparr=="S")
start=np.array([startx[0],starty[0]])
print(start)
startdir=1
#Find target
[targetx,targety]=np.where(maparr=="E")
target=np.array([targetx[0],targety[0]])
print(target)
```

```python
visited=np.full((dimx,dimy),-1)
wallvisited=np.full((dimx,dimy),-1)
locs=np.array([start])
visited[start[0],start[1]]=0
walls=[]
wallcosts=[]
shortcuts=[]
track=[start]
while len(locs)>0 and visited[target[0],target[1]]==-1:
    #print('New round')
    newlocs=[]
    #newdirs=[]
    #newcosts=[]
    for loc in locs:
        cost=visited[loc[0],loc[1]]
        #print('Location', loc,'Cost',cost)
        for dir in directions:
           # print ('direction', dir, 'target', loc+dir)
            #if loc[0]+dir[0]>=0 and loc[1]+dir[1]>=0 and loc[0]+dir[0]<dimx and loc[1]+dir[1]<dimy:
            if maparr[loc[0]+dir[0],loc[1]+dir[1]]!='#': 
                if visited[loc[0]+dir[0],loc[1]+dir[1]]==-1:
                    #print('Success')
                    newlocs.append(loc+dir)
                    track.append(loc+dir)
                    visited[loc[0]+dir[0],loc[1]+dir[1]]=cost+1
            elif all(loc+dir>=[0,0]) and all(loc+dir<[dimx,dimy]): #wall
                if wallvisited[loc[0]+dir[0],loc[1]+dir[1]]==-1:
                    wallvisited[loc[0]+dir[0],loc[1]+dir[1]]=cost+1
                elif wallvisited[loc[0]+dir[0],loc[1]+dir[1]]<cost-1: #shortcut
                    shortcuts.append(cost-wallvisited[loc[0]+dir[0],loc[1]+dir[1]]-1)
                    walls.append(loc+dir)
                    wallcosts.append(cost+1)
        locs=np.array(np.unique(newlocs,axis=0))
        #locs=np.array(newlocs)
#walls=np.array(np.unique(walls,axis=0))
track=np.array(track)
print(visited)
print(track)
print('Final',visited[target[0],target[1]])
basecost=visited[target[0],target[1]]
basevisited=visited.copy()
basewallvisited=wallvisited.copy()
shortcuts=np.array(shortcuts)
print(shortcuts)
print(np.unique(shortcuts,return_counts=True))
```
