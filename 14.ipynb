```python
import numpy as np
import re
```

```python
positions=[]
velocities=[]
with open('input_14.txt') as f:
    for line in f:
        m=re.match('p=([0-9]+),([0-9]+) v=([0-9-]+),([0-9-]+)',line)
        positions.append([int(m.group(1)),int(m.group(2))])
        velocities.append([int(m.group(3)),int(m.group(4))])
print(positions, velocities)
positions=np.array(positions)
velocities=np.array(velocities)
```

```python
dim=[101,103]
steps=100
dims=np.tile(np.array(dim),(positions.shape[0],1))
newpos=positions+steps*velocities
newpos = np.mod(newpos,dims)
print(newpos)
```

```python
#Count em
#print(dim[0]//2,dim[1]//2)
count=np.zeros(4,dtype=int)
for i in range(newpos.shape[0]):
    #print(newpos[i])
    if newpos[i,0]<dim[0]//2:
        if newpos[i,1]<dim[1]//2:
#            print('q0')
            count[0] += 1
        elif newpos[i,1]>dim[1]//2:
#            print('q1')
            count[1] += 1
    elif newpos[i,0]>dim[0]//2:
        if newpos[i,1]<dim[1]//2:
#            print('q2')
            count[2] += 1
        elif newpos[i,1]>dim[1]//2:
#            print('q3')
            count[3] += 1
print(count, np.prod(count))
#print(newpos[:,0]<dim[0]//2)
#q1=np.count_nonzero(newpos[:,0]<dim[0]//2 and newpos[:,1]<dim[1]//2))
#print(q1)
```

```python
#part b
dim=[101,103]
steps=100
dims=np.tile(np.array(dim),(positions.shape[0],1))
for step in range(10000):
    #print('Step', step)
    newpos=positions+step*velocities
    newpos = np.mod(newpos,dims)
    count=np.zeros(dim,dtype=int)
    for i in range(newpos.shape[0]):
        #print(newpos[i])
        count[newpos[i][0],newpos[i][1]]=1
    found=0
    for i in range(dim[1]):
        if re.search('1{6,}',''.join((count[:,i].astype(str)))):
            found=1
            break
    if found:
        for i in range(dim[1]):
            print(step,''.join((count[:,i].astype(str))).replace('0',' '))
    print(step,np.var(newpos[:,0]),np.var(newpos[:,1]))


#print(newpos)
```
