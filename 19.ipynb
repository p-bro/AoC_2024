```python
import numpy as np
```

```python
patterns=[]
with open('input_19.txt') as f:
    for line in f:
        if len(line)<2: #blank lines separate rules from updates
            break
        towels=line.strip().split(', ')
    for line in f:
        patterns.append(line.strip())
print(towels)
print(patterns)
```

```python
patterncount=0
for pattern in patterns:
    #print('Pattern',pattern)
    towellist=[]
    poslist=[0]
    numpos=[1]
    posseen=np.zeros(len(pattern)+1)
    complete=0
    while True: #not complete: #pos<=len(pattern):
        newpos=[]
        newcounts=[]
        for pos,num in zip(poslist,numpos):
            #print('Position',pos)
            for towel in towels:
                if pos+len(towel)>len(pattern):
                    #print("Towel too long:",towel)
                    continue
                match=1
                for numstripe,stripe in enumerate(list(towel)):
                    if pattern[pos+numstripe]!=stripe:
                        #print('Towel no match:', towel)
                        match=0
                        break
                if match:
                    #if posseen[pos+len(list(towel))]:
                        #print('Pos already visited:',towel,pos+len(list(towel)))
                        #continue
                    #else:
                        if pos+len(list(towel))==len(pattern):
                            complete+=num
                            print('Pattern complete', num,pattern)
                        else:
                            newpos.append(pos+len(list(towel)))
                            newcounts.append(num)
                        #print('Towel match:',towel,pos,pattern)
                        #    posseen[pos+len(list(towel))]=1

        if len(newpos):
            poslist,invpos=np.unique(np.array(newpos),return_inverse=True)
            numpos=np.zeros_like(poslist)
            #print('New positions',newpos,newcounts)
            for i,index in enumerate(invpos):
                numpos[index]+=newcounts[i]
            #print('New positions unique',poslist,numpos)
        else:
            print('pattern impossible',pattern)
            break
    if complete:
        patterncount+=complete
print('Completed patterns',patterncount)    
        

```
