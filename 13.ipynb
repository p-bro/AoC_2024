```python
import numpy as np
```

```python
A=[]
B=[]
prize=[]
cost=np.array([3,1])
machinecount = 0
linecount=0
with open('input_13.txt') as f:
    for line in f:
        linecount+=1
        text=list(line.strip())
#        print(text,text[7])
        if len(text)>0:
            if text[7]=='A':
                A.append([int(text[12]+text[13]),int(text[18]+text[19])])
            elif text[7]=='B':
                B.append([int(text[12]+text[13]),int(text[18]+text[19])])
            else: #prize
                prizeline=line.strip().split()
                #print(prizeline,prizeline[1][2:-1],prizeline[2][2:])
                prize.append([int(prizeline[1][2:-1])+10000000000000,int(prizeline[2][2:])+10000000000000])
                machinecount+=1
print(A,B,prize)
print(len(prize),machinecount,linecount)
```

```python
totaltok=0
for machine in range(machinecount):
    #print(A[machine],B[machine],prize[machine])
    #solve x1*a1+x2*b1=prize1, x1*a2+x2*b2=prize2
    mat=np.array([A[machine],B[machine]]).transpose()
    #print(mat,np.array(prize[machine]))
    sol=np.linalg.solve(mat,np.array(prize[machine]))
    intsol=np.round(sol,0).astype(int)
    #print(sol,intsol,np.matmul(mat,intsol),np.array(prize[machine]))
    if all(np.matmul(mat,intsol)==np.array(prize[machine])) and all(intsol>=0):
        tokens=np.dot(intsol,cost).astype(int)
        totaltok+=tokens
        print('Solution',intsol,tokens,totaltok)
    else: 
        print('No',sol,intsol,np.matmul(mat,intsol),np.array(prize[machine]))
print(totaltok)
```
