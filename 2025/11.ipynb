```python
import numpy as np
import networkx as nx
```

```python
G =nx.DiGraph()
with open('input_11.txt') as f:
    for line in f:
        nodeline=line.strip().split(':')
        G.add_node(nodeline[0])
        dests=nodeline[1].strip().split()
        for dest in dests:
            G.add_edge(nodeline[0],dest)
visitarr=np.zeros(G.number_of_nodes(),dtype=int)
for node in list(G.nodes):
    G.nodes[node]['visits']=np.zeros(G.number_of_nodes()+1,dtype=int)
```

```python
countpaths=0
for path in nx.all_simple_paths(G, source="you", target="out"):
    countpaths+=1
    print(path)
print(countpaths)
```
