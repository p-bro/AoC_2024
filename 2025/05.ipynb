```python
import numpy as np
```

```python
ranges=[]
products=[]
with open('input_05.txt') as f:
    for line in f:
        if len(line)<2: #blank lines separate rules from updates
            break
        ranges.append(line.strip().split('-'))
    for line in f:
        products.append(int(line.strip()))
prodarr=np.array(products,dtype=int)
print(len(ranges),ranges)
numranges=len(ranges)
idrange=np.zeros([len(ranges),2],dtype=int)
for i in range(numranges):
    idrange[i,0]=ranges[i][0]
    idrange[i,1]=ranges[i][1]
print(idrange)
print(prodarr)
```

```python
#a
numspoilt=0
numfresh=0
for prod in prodarr:
    valid = False
    for idr in range(numranges):
        if prod > idrange[idr,0] and prod <= idrange[idr,1]:
            valid = True
            print (prod, "valid in range", idrange[idr,:])
            numfresh+=1
            break
    if not valid:
        numspoilt += 1
        print (prod, "spoilt prod no.", numspoilt)
print ("spoilt", numspoilt, "fresh",numfresh)
```

```python
#b
freshids=0
for idra in range(numranges):
    if idrange[idra,0]<0:
        continue
    freshids += idrange[idra,1]+1-idrange[idra,0]
    print("range added",idrange[idra,:],idrange[idra,1]+1-idrange[idra,0])
    for idrb in range(idra):
        #left boundary in interval
        if idrange[idra,0]>=idrange[idrb,0] and idrange[idra,0]<=idrange[idrb,1]:
            #right boundary in interval
            if idrange[idra,1]<=idrange[idrb,1]: #remove completely
                print("Full overlap", idrange[idra,:],idrange[idrb,:])
                print("removed",idrange[idra,:],idrange[idra,1]+1-idrange[idra,0])
                freshids -= idrange[idra,1]+1-idrange[idra,0]
                idrange[idra,:]=-1
                break
            else:
                print("Intervals overlap partially a", idrange[idra,:],idrange[idrb,:], "removed", idrange[idrb,1]+1-idrange[idra,0])
                freshids -= idrange[idrb,1]+1-idrange[idra,0] #remove overlap
                idrange[idrb,1]=idrange[idra,0]-1
                print("Adjusted interval",idrange[idrb,:])
        elif idrange[idra,1]>=idrange[idrb,0] and idrange[idra,1]<idrange[idrb,1]:
        #right boundary in interval (but not left)
            freshids -= idrange[idra,1]+1-idrange[idrb,0] #remove overlap
            print("Intervals overlap partially b", idrange[idra,:],idrange[idrb,:], "removed", idrange[idra,1]+1-idrange[idrb,0])
            idrange[idrb,0]=idrange[idra,1]+1
            print("Adjusted interval",idrange[idrb,:])
        elif idrange[idrb,0]>=idrange[idra,0] and idrange[idrb,0]<=idrange[idra,1]:
            #other interval in new interval
            print("Full overlap b", idrange[idra,:],idrange[idrb,:])
            print("removed",idrange[idrb,:],idrange[idrb,1]+1-idrange[idrb,0])

            freshids -= idrange[idrb,1]+1-idrange[idrb,0]
            idrange[idrb,:]=-1
        else:
            print("No overlap between",idrange[idra,:],idrange[idrb,:])
print ("Fresh ids",freshids)
```
