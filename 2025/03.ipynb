```python
import numpy as np
```

```python
batts=[]
linecount = 0
with open('input_03.txt') as f:
    for line in f:
        batts.append(list(map(int,line.strip())))
battarr=np.array(batts)
numbanks=battarr.shape[0]
numbatts=battarr.shape[1]
print(numbanks,numbatts,battarr)
```

```python
jolts=0
for bank in range(numbanks):
    tens=np.max(battarr[bank,:-1])
    postens=np.argmax((battarr[bank,:-1]))
    ones=np.max(battarr[bank,postens+1:])
    jolt=tens*10+ones
    print(postens,tens,ones,tens*10+ones)
    jolts += jolt
print (jolts)
```

```python
#b 12 digits
jolts=0
digits=12
for bank in range(numbanks):
    start=0
    jolt=0
    print(battarr[bank,:])
    for digit in range(digits):
        #print(digit,start,digits-digit,battarr[bank,start:numbatts-(digits-digit-1)])
        digval=np.max(battarr[bank,start:numbatts-(digits-digit-1)])
        #print(digit,start,digits-digit,battarr[bank,start:numbatts-(digits-digit-1)],digval)
        start=np.argmax(battarr[bank,start:numbatts-(digits-digit-1)])+1+start
        jolt = jolt*10 + digval
    print(battarr[bank,:],jolt)
    jolts += jolt
print (jolts)
```
