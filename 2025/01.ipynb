```python
import numpy as np
import pandas as pd
```

```python
comb=[]
linecount = 0
with open('input_01a.txt') as f:
    for line in f:
        entry=line.strip()
        if (entry[0]=='R'):
            advance=int(entry[1:])
        else:
            advance=-int(entry[1:])
        comb.append(advance)
combarr=np.array(comb)
print(combarr)
```

```python
pos=50
zeros=0
for i in combarr:
    pos = (pos+i)%100
    if (pos==0):
        zeros+=1
    print(i, pos)
print(zeros)
```

```python
#Method b
pos=50
zeros=0
oldzeros=0
for i in combarr:
    if (pos==0 and i<0):
        zeros-=1
    pos = (pos+i)
    if (pos==0):
        zeros+=1
    #passing zero
    zeros += np.abs((pos % 100-pos)//100)
    pos = pos % 100
    print(i, pos, zeros-oldzeros)
    oldzeros=zeros
print(zeros)
```

```python
#brute force b
#Method b
pos=50
zeros=0
oldzeros=0
for i in combarr:
    for j in range(np.abs(i)):
        if (i>0):
            pos = (pos+1)%100
        else:
            pos = (pos-1)%100    
        if pos==0:
            zeros+=1
    print(i, pos, zeros-oldzeros)
    oldzeros=zeros
print(zeros)
```
