```python
maxarea=0
for i in range(len(vertices)):
    vertex1=redarr[vertices[i],:]
    for j in range(i):
        vertex2=redarr[vertices[j],:]
        area=(np.abs(vertex1[0]-vertex2[0])+1)*(np.abs(vertex1[1]-vertex2[1])+1)
        if area>maxarea:
            maxarea=area
print(maxarea)
```
