```python
import numpy as np
import pandas as pd
```

```python
ranges=[]
with open('input_02a.txt') as f:
    for line in f:
        ranges=list(line.strip().split(','))
print(len(ranges),ranges)
numranges=len(ranges)
idrange=np.zeros([len(ranges),2],dtype=int)
for i in range(numranges):
    bounds=ranges[i].split('-')
    idrange[i,0]=bounds[0]
    idrange[i,1]=bounds[1]
print(idrange)
#    idra
```

```python
invalid=0
for i in range(numranges):
    lo = idrange[i][0]
    hi = idrange[i][1]
    digits = len(str(lo))
    if digits % 2 == 1:
        lo = int(10**digits)
        digits += 1
    halfdigs=digits//2
    print(lo,hi,halfdigs)
    if lo > hi:
        continue
    half = lo//(10**(halfdigs))
    while half*10**(halfdigs)+half <= hi:
        if half*10**(halfdigs)+half >= lo:
            print(half*10**(halfdigs)+half, "invalid")
            invalid += half*10**(halfdigs)+half
        half += 1
        if half % (10**halfdigs) == 0:
            halfdigs += 1
            half *=10
print (invalid)
```
