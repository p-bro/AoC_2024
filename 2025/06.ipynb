```python
import numpy as np
```

```python
# Define addition operation function
def add(a, b):
    return a+b

# Define subtraction operation function
def sub(a, b):
    return a-b

# Define multiplication operation function
def mul(a, b):
    return a*b

# Define division operation function
def div(a, b):
    return a/b
```

```python
data=[]
with open('test_06.txt') as f:
    for line in f:
        data.append(line.strip().split())
print(data,len(data))
dataarr = np.array(data[:-1],dtype=int)
oparr = np.array(data[-1])
numeqn = dataarr.shape[1]
numops = dataarr.shape[0]
print(dataarr,oparr,numeqn,numops)
```

```python
chksum=0
for eqn in range(numeqn):
    value = dataarr[0,eqn]
    for item in range(1,numops):
        if oparr[eqn]=="+":
            value += dataarr[item,eqn]
        else: # mult
            value *= dataarr[item,eqn]
    print(value)
    chksum+=value
print(chksum)
```
