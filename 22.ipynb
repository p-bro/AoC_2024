```python
import numpy as np
```

```python
seeds=[]
with open('input_22.txt') as f:
    for line in f:
        seeds.append(line.strip())
seeds=np.array(seeds).astype(int)
print(seeds)
```

```python
def mix(a,b):
    return a ^ b

def prune(a):
    return a % 16777216

def evolve(a):
    a = prune(mix(a,a*64))
    a = prune(mix(a,a // 32))
    a = prune(mix(a,a*2048))
    return a
```

```python
secrets=seeds.copy()
lastdigs=np.array([seeds % 10])
for i in range(2000):
    secrets=evolve(secrets)
    lastdigs=np.concatenate((lastdigs,np.array([secrets%10])))
for seed,secret in zip(seeds,secrets):
    print (seed, secret)
print(sum(secrets))
```

```python
print(lastdigs.shape)
diffs=np.empty_like(lastdigs)
for step in range(1,lastdigs.shape[0]):
    diffs[step,:]=lastdigs[step,:]-lastdigs[step-1,:]
```

```python
stencils=[]
for i in range(-9,10):
    for j in range(-9,10):
        if i+j >= 10 or i+j<=-10:
            continue
        for k in range (-9,10):
            if j+k>=10 or j+k <= -10 or i+j+k >=10 or i+j+k <= -10:
                continue
            for m in range (-9,10):
                if k+m >=10 or k+m <= -10 or j+k+m >=10 or j+k+m <=-10 or i+j+k+m >=10 or i+j+k+m <=-10:
                    continue
                stencils.append([i,j,k,m])
stencils=np.array(stencils)
```

```python
print(stencils)
print(stencils.shape)
```

```python
bought=np.zeros(stencils.shape[0],dtype=int)
for count,stencil in enumerate(stencils):
    print(stencil)
    for monkey in range(len(seeds)):
        for step in range(4,2001):
            if all(diffs[step-3:step+1,monkey]==stencil[:]):
                bought[count]+=lastdigs[step,monkey]
                print('Bought', lastdigs[step,monkey], 'from', monkey)
                break
```
